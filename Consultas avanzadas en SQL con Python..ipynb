```python
cur.execute("""
    SELECT Sales, AVG(Last_evaluation)
    FROM Detalle
    GROUP BY Sales;
""")
res = cur.fetchall()
res
```

```python
cur.execute("""
    SELECT Sales, Salary, AVG(Last_evaluation)
    FROM Detalle
    GROUP BY Sales, Salary
""")
res = cur.fetchall()
res
```

```python
cur.execute("""
    SELECT Sales, AVG(Work_accident)
    FROM Detalle
    GROUP BY Sales
    HAVING COUNT(Work_accident) > 0.15;
""")
res= cur.fetchall()
res
```
